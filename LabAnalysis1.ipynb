##Analysis of length change measurements
1. Calculate the mean ($\mu$), standard deviation ($\sigma$), and standard error of the mean for the set of measurements before you changed the pendulum length and after you changed the pendulum length. Use 1/(N-1) to calculate the standard deviation. 
1. Using $L = g \left(\frac{T}{2\pi}\right)^2$, calculate the length of the pendulum before and after the change. IE if the period before the change is $T_1 \pm \Delta T_1$, calculate $L_1 \pm \Delta L_1$. You can do this by applying the formula for $T_1$, $T_1 + \Delta T_1$ and $T_1 - \Delta T_1$ or by taylor expansion/error propogation. Should you use the standard deviation or the standard error for $\Delta T_1$. Do the same calculation to find $L_2 \pm \Delta L_2$. 
1. Please make a histogram showing the two sets of measurements on the same axis. 
  1. Start with 20 bins spanning 3 standard deviations below the lowest mean to 3 standard deviations above the highest mean. If the bins are underfilled (i.e. max bin has <= 3 elements), then adjust the number of bins down. If all the data is in only a few bins, then adjust the number of bins up. 
  1. Make the before and after histograms in different colors; use the options histtype='stepfilled' and alpha = 0.75 so that overlapping bins can be seen clearly.

```python
#1
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

def mean(input):
  N = np.size(input)
  total = sum(input)
  mean = total/N
  return(mean)

mean_pre = mean(pre_change)
mean_post = mean(post_change)

def sigma(input):
  N = np.size(input)
  u = mean(input)
  variance = []
  for i in input:
    variance.append((i-u)**2)
  sum_variance = sum(variance)

  sigma = np.sqrt((sum_variance)/(N-1))
  return(sigma)

sigma_pre = sigma(pre_change)
sigma_post = sigma(post_change)


def SEM(input):
  SD = sigma(input)
  N = np.size(input)
  SEM = (SD/np.sqrt(N))
  return(SEM)

SEM_pre = SEM(pre_change)
SEM_post = SEM(post_change)


from tabulate import tabulate
table_2 = {'Method':['Pre-Change', 'Post-Change'], 'Mean(seconds)': [(mean_pre), (mean_post)], 'Standard Deviation': [(sigma_pre), (sigma_post)], 'Standard Error': [(SEM_pre), (SEM_post)]}
print(tabulate(table_2, headers = 'keys'))

#2
#Using  L=g(T2)2 , calculate the length of the pendulum before and after the change. 
#IE if the period before the change is  T1T1 , calculate  L1L1 . 
#You can do this by applying the formula for  T1 ,  T1+T1  and  T1T1  or by taylor expansion/error propogation. 
#Should you use the standard deviation or the standard error for  T1 . 
#Do the same calculation to find  L2L2 

def T_to_L(Tmean):
  L = 9.81*((Tmean/2*np.pi)**2)
  return(L)

L1 = T_to_L(mean_pre)
L2 = T_to_L(mean_post)

#using error propagation and standard deviation, (dL/L)=2(dT/T)
def Lvariance(Tmean, Tstderr):
  Lmean = 9.81*((Tmean/2*np.pi)**2)
  Lstderr = Lmean*2*(Tstderr/Tmean)
  return(Lstderr) 

Lstderr1 = Lvariance(mean_pre, SEM_pre)
Lstderr2 = Lvariance(mean_post, SEM_post)

print('Part 2:')
print(f'L1(pre-change) = {L1} +- {Lstderr1}cm')
print(f'L2(post-change) = {L2} +- {Lstderr2}cm')


#3
#Please make a histogram showing the two sets of measurements on the same axis.
#Start with 20 bins spanning 3 standard deviations below the lowest mean to 3 standard deviations above the highest mean. If the bins are underfilled (i.e. max bin has <= 3 elements), then adjust the number of bins down. If all the data is in only a few bins, then adjust the number of bins up.
#Make the before and after histograms in different colors; use the options histtype='stepfilled' and alpha = 0.75 so that overlapping bins can be seen clearly.
#Based on this histogram can you tell that when you changed the length, you also changed the period?
lowerbound = mean_post - 3*sigma_pre
upperbound = mean_pre + 3*sigma_pre 
r = np.linspace(lowerbound, upperbound, 20)

plt.hist([pre_change], bins=r, label = 'Pre-Change', alpha = 0.75)
plt.hist([post_change], bins=r, label='Post-Change', alpha = 0.75)
plt.legend(loc = 'upper right')
plt.xlabel('Measured Period (seconds)')
plt.ylabel('Count of Measured Periods')
plt.title('Spread of Periods Among Data Collected during Pre- and Post- Length Change Meansurements')

plt.show()
print("ANSWER: Yes, I can tell(by seperating the two measurements into two different colors) that the means of the period before and after the length change are different. Because there is also a 'valley' or 'gap' inbetween the spread of data between the two different sets of measurements, I can guess that the means of each data set probably fall outside of 3sigma's of each other's measurements.")


#4
together = np.concatenate((pre_change, post_change))
plt.hist([together], bins=r, alpha = 0.75)
plt.xlabel('Measured Period (seconds)')
plt.ylabel('Count of Measured Periods')
plt.title('GROUPED: Spread of All Periods Measured across Pre- and Post- Length Change Trials')
plt.show()
print("ANSWER: Yes, I can still tell that there are two lengths being measured. There are two distinct maxima on the histogram and a 'valley' in between.")

```
